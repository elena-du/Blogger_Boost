```python
tokens = [sub.split() for subl in  text0_l for sub in subl] 
```

```python
embedding_dim=300
```

```python
w_v = Word2Vec(tokens, size=embedding_dim, window=5, 
               min_count=5, negative=15, iter=3, workers=multiprocessing.cpu_count())
```

```python
word_vectors = w_v.wv
```

```python
result=word_vectors.similar_by_word('blog')
```

```python
print('Most similar to "blog": ', result[:3])
```
